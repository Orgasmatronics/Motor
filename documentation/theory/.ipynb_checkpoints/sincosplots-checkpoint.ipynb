```python
%matplotlib inline   
from pylab import *  

```

```python
t = linspace(0,4.*pi,1000)
omega = 1.0
x = cos(omega*t)
y = sin(omega*t)
plot(t,y,t,x,linewidth = 2)
grid()
axis([0,4.*pi,-1.2,1.2])
xlabel('time',fontsize = 16)
ylabel('x,y',fontsize = 16)
```

```python
figure(figsize = [4,4])
plot(x[0:600],y[0:600],linewidth = 2)
axis([-1.2,1.2,-1.2,1.2])
grid()
ylabel('y',fontsize = 16)
xlabel('x',fontsize = 16)
z = x + 1.j*y
```

```python
plot(angle(z),linewidth = 2)
grid()
```
