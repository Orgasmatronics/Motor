We start with a simple harmonic oscillator described by 

$\ddot{x} = - \omega_0^2 x - \frac{\omega_0}{Q}\dot{x}$

Force is a feedback signal that turns on in the positive direction only for positive velocities only(i.e. same direction).  So when $\dot{x} \leq 0$ the force has to be zero, and it is a constant $F_{feedback}$ for all $\dot{x} > 0$.  That's not continuous or differentialble, and we need that for the ODE solver to work.  So instead we use a hyperbolic tangent to get a smooth function.  This will leave a residue of force of the wrong sign for all negative velocities but they can be exponentially small compared to the positive force so it's ok.  

$F = \frac{1}{2}\left(\tanh\left(\frac{\dot{x}- \epsilon\omega_0}{\epsilon\omega_0}\right) + 1\right)F_{feedback}$, 

where $\epsilon << 1$ is turn on rate normalized to the resonant frequency.  This function is plotted below.

```python
x = linspace(-2,2,1000)
y = 0.5*(tanh((x - 0.04)/0.04) + 1.0)
plot(x,y, linewidth = 2)
axis([-2,2,-.2,1.2])
grid()
```

$Z_0 = \sqrt{km}$

$\omega_0 = \sqrt{\frac{k}{m}}$

$m = \frac{Z_0}{\omega_0}$, $\frac{F_{feedback}}{m} = \frac{F_{feedback}\omega_0}{Z_0}$

$\ddot{x} = - \omega_0^2 x - \frac{\omega_0}{Q}\dot{x} +  \frac{\omega_0}{Z_0} \frac{1}{2}\left(\tanh\left(\frac{\dot{x}- \epsilon\omega_0}{\epsilon\omega_0}\right) + 1\right)F_{feedback} $

In the following equation I'm setting m to 1.0:


```python
def feedbackSHO1(y, t, omega_0,Q,F_feedback,epsilon):   #differential operator for pendulum, with damping, no drive, and zero current bias
    F = F_feedback*0.5*(tanh((y[1] - epsilon*omega_0)/(epsilon*omega_0)) + 1.0)
    return [y[1], -omega_0**2*y[0]  - (omega_0/Q)*y[1] + F]
```

```python
f_0 = 1.0 #frequency in Hz
omega_0 = 2*pi*f_0
Q = 5
F_feedback = 50.0
epsilon = 0.001
t = linspace(0,20.,8000)
odeoutput = odeint(feedbackSHO1,[1.,0.], t,(omega_0,Q,F_feedback,epsilon))
x = odeoutput[:,0]
xdot = odeoutput[:,1]

plot(t,xdot/omega_0,'o',t,22.0*(t - 19.6),t,29.0*(t - 19.6),linewidth = 2)
axis([19.5,19.7,-2,2])
grid()
```

```python
plot(x,xdot/omega_0)
grid()
```

```python
def feedbackSHO2(y, t, omega_0,Q,F_feedback,epsilon,delta):   #differential operator for pendulum, with damping, no drive, and zero current bias
    F = F_feedback*0.5*(tanh((y[1] - epsilon*omega_0)/(epsilon*omega_0)) + 1.0)
    return [y[1], -omega_0**2*y[0] - delta*y[0]**3  - (omega_0/Q)*y[1] + F] #add nonlinearity to play with the concept
```

Note also that the accurate picture of the mass on spring with feedback drive includes the $1/r^3$ dropoff of the field and hence the force.  Adding a cubic term to the force adds a inverse square term to the potential, but only for the positive going velocities.

```python
def feedbackSHO3(y, t, omega_0,Q,F_feedback,epsilon,m):   #differential operator for pendulum, with damping, no drive, and zero current bias
    F = F_feedback*0.5*(tanh((y[1] - epsilon*omega_0)/(epsilon*omega_0)) + 1.0) #*(y[0] - 1.0)**(-3.0)
    return [y[1], -omega_0**2*y[0] - (omega_0/Q)*y[1] + F/m] #add nonlinearity to play with the concept
```

```python
f_0 = 20.0 #frequency in Hz
omega_0 = 2*pi*f_0
m = mass.magnitude #in kg
Q = 5
F_feedback = 3.0 #newtons
epsilon = 0.001
t = linspace(0,1.,8000)
odeoutput = odeint(feedbackSHO3,[0.01,0.], t,(omega_0,Q,F_feedback,epsilon,m))
x = odeoutput[:,0]
xdot = odeoutput[:,1]
plot(x,xdot/omega_0)
```
